# SVM lineal

En Scikit Learn pueden encontrarse tres implementaciones distintas del algoritmo Suport Vector Machine:

Las clases `sklearn.svm.SVC` y `sklearn.svm.NuSVC` permiten crear modelos SVM de clasificaci칩n empleando kernel lineal, polinomial, radial o sigmoide. La diferencia es que `SVC` controla la regularizaci칩n a trav칠s del hiperpar치metro C, mientras que `NuSVC` lo hace con el n칰mero m치ximo de vectores soporte permitidos.

La clase `sklearn.svm.LinearSVC` permite ajustar modelos SVM con kernel lineal. Es similar a SVC cuando el par치metro `kernel='linear'`, pero utiliza un algoritmo m치s r치pido.

Las mismas implementaciones est치n disponibles para regresi칩n en las clases: `sklearn.svm.SVR`, `sklearn.svm.NuSVR` y `sklearn.svm.LinearSVR`.

Se ajusta primero un modelo SVM con kernel lineal y despu칠s uno con kernel radial, y se compara la capacidad de cada uno para clasificar correctamente las observaciones.

###  Divisi칩n de los datos en train y test

```python
X = datos.drop(columns = 'y')
y = datos['y']
```

```python
X.X1.values.reshape(4, -1)
```

```python
y.values.reshape(-1,1)
```

```python

X_train, X_test, y_train, y_test = train_test_split(
                                        X,
                                        y.values.reshape(-1,1),
                                        train_size   = 0.8,
                                        random_state = 1234,
                                        shuffle      = True
                                    )
```

```python
# Creaci칩n del modelo SVM lineal
modelo = SVC(C = 100, kernel = 'linear', random_state=123)
modelo.fit(X_train, y_train)
```

Se calcula el porcentaje de aciertos que tiene el modelo al predecir las observaciones de test (accuracy).



```python
# Predicciones test
predicciones = modelo.predict(X_test)
predicciones
```

# SVM radial

Se repite el ajuste del modelo, esta vez empleando un kernel radial y utilizando `validaci칩n cruzada` para identificar el valor 칩ptimo de penalizaci칩n` C.`

```python
 np.logspace(-5, 7, 20)
```

```python
# Grid de hiperpar치metros
# ==============================================================================
param_grid = {'C': np.logspace(-5, 7, 20),
              # 'gamma': 
              # 'kernel':
              }

# B칰squeda por validaci칩n cruzada
# ==============================================================================
grid = GridSearchCV(
        estimator  = SVC(kernel= "rbf", gamma='scale'),
        param_grid = param_grid,
        scoring    = 'accuracy',
        n_jobs     = -1,
        cv         = 3, 
        verbose    = 0,
        return_train_score = True
      )
```

```python
# Se asigna el resultado a _ para que no se imprima por pantalla
X = X_train
y = y_train

```

```python
_ = grid.fit(X,y)
```

```python

# Resultados del grid
resultados = pd.DataFrame(grid.cv_results_)
resultados.filter(regex = '(param.*|mean_t|std_t)').drop(columns = 'params').sort_values('mean_test_score', ascending = False).head(5)
```

```python
# Grid de hiperpar치metros
param_grid = {'C': np.logspace(-5, 7, 20), 'kernel': ['rbf', 'linear']}

# B칰squeda por validaci칩n cruzada
grid = GridSearchCV(
        estimator  = SVC(gamma='scale'),
        param_grid = param_grid,
        scoring    = 'accuracy',
        n_jobs     = -1,
        cv         = 3, 
        verbose    = 0,
        return_train_score = True
      )

# Se asigna el resultado a _ para que no se imprima por pantalla
_ = grid.fit(X = X_train, y = y_train)

# Resultados del grid
resultados = pd.DataFrame(grid.cv_results_)
resultados.filter(regex = '(param.*|mean_t|std_t)')\
    .drop(columns = 'params')\
    .sort_values('mean_test_score', ascending = False) \
    .head(5)
```

```python
# Mejores hiperpar치metros por validaci칩n cruzada
print("----------------------------------------")
print("Mejores hiperpar치metros encontrados (cv)")
print("----------------------------------------")
print(grid.best_params_, ":", grid.best_score_, grid.scoring)

modelo = grid.best_estimator_
```

```python
# Representaci칩n gr치fica de los l칤mites de clasificaci칩n
# Grid de valores
x = np.linspace(np.min(X_train.X1), np.max(X_train.X1), 50)
y = np.linspace(np.min(X_train.X2), np.max(X_train.X2), 50)
Y, X = np.meshgrid(y, x)
grid = np.vstack([X.ravel(), Y.ravel()]).T

# Predicci칩n valores grid
pred_grid = modelo.predict(grid)

fig, ax = plt.subplots(figsize=(6,4))
ax.scatter(grid[:,0], grid[:,1], c=pred_grid, alpha = 0.2)
ax.scatter(X_train.X1, X_train.X2, c=y_train, alpha = 1)

# Vectores soporte
ax.scatter(
    modelo.support_vectors_[:, 0],
    modelo.support_vectors_[:, 1],
    s=200, linewidth=1,
    facecolors='none', edgecolors='black'
)

# Hiperplano de separaci칩n
ax.contour(
    X,
    Y,
    modelo.decision_function(grid).reshape(X.shape),
    colors='k',
    levels=[0],
    alpha=0.5,
    linestyles='-'
)

ax.set_title("Resultados clasificaci칩n SVM radial");
```

```python
# Representaci칩n gr치fica utilizando plot_decision_regions() de mlxtend
fig, ax = plt.subplots(figsize=(6,4))
plot_decision_regions(
    X = X_train.to_numpy(),
    y = y_train.flatten(),
    clf = modelo,
    ax = ax
)
ax.set_title("Resultados clasificaci칩n SVM radial");
```

```python
# Predicciones test
predicciones = modelo.predict(X_test)
```

```python
# Accuracy de test del modelo 
accuracy = accuracy_score(
            y_true    = y_test,
            y_pred    = predicciones,
            normalize = True
           )
print("")
print(f"El accuracy de test es: {100*accuracy}%")
```

```python
# Matriz de confusi칩n de las predicciones de test
confusion_matrix = pd.crosstab(
    y_test.ravel(),
    predicciones,
    rownames=['Real'],
    colnames=['Predicci칩n']
)
confusion_matrix
```

Conclusi칩n

Con un modelo SVM de kernel radial se consigue clasificar correctamente el 80% de las observaciones de test. Se podr칤a intentar mejorar a칰n m치s el modelo optimizando el valor del hiperpar치metro gamma o utilizando otro tipo de kernel.
