# t-SNE  - Reducción de Dimensiones 
### `skills.tech`

La API de Scikit-learn proporciona la clase TSNE para visualizar datos con el método T-SNE. En este tutorial, aprenderemos brevemente cómo ajustar y visualizar datos con TSNE en Python. Los tutoriales cubren:

* Ajuste y visualización de TSNE del conjunto de datos de iris
* Ajuste y visualización de TSNE del conjunto de datos MNIST
* Listado de código fuente

```python
from sklearn.manifold import TSNE
from keras.datasets import mnist  #recuerda instalar keras y tensorflow con pip
from sklearn.datasets import load_iris
from numpy import reshape
import seaborn as sns
import pandas as pd
```

```python
iris = load_iris()
x = iris.data
y = iris.target
```

```python
x
```

```python
y
```

Definiremos el modelo usando la clase TSNE, aquí el parámetro n_components define el número de dimensiones objetivo. El 'verbose = 1' muestra los datos de registro para que podamos verificarlos.

```python
tsne = TSNE(n_components=2, verbose=0, random_state=100)
z = tsne.fit_transform(x)
z[:10,:]
```

```python
tsne.fit(x[:-1])
```

```python
tsne.transform(x[:-1])
```

```python
tsne.fit_transform(x[:-1])
```

A continuación, visualizaremos el resultado en una gráfica. Recopilaremos los datos del componente de salida en un marco de datos, luego usaremos el diagrama de dispersión () de la biblioteca 'seaborn' para trazar los datos. En la paleta de colores del diagrama de dispersión, estableceremos 3 porque hay 3 categorías de tipos en los datos de la etiqueta.

```python
df = pd.DataFrame()
df["y"] = y
df["comp-1"] = z[:,0]
df["comp-2"] = z[:,1]
df["x1"] = x[:,0]
df["x2"] = x[:,1]

sns.scatterplot(x="x1", y="x2", hue=df.y.tolist(),
                palette=sns.color_palette("hls", 3),
                data=df).set(title="Iris data projection")
```

```python
df = pd.DataFrame()
df["y"] = y
df["comp-1"] = z[:,0]
df["comp-2"] = z[:,1]
sns.scatterplot(x="comp-1", y="comp-2", hue=df.y.tolist(),
                palette=sns.color_palette("hls", 3),
                data=df).set(title="Iris data T-SNE projection")
```

### MNIST dataset TSNE fitting and visualizing


A continuación, aplicaremos el mismo método al conjunto de datos más grande. El conjunto de datos de dígitos manuscritos MNIST funciona bien para este propósito y podemos usar los datos MNIST de la API de Keras. Extraemos solo una parte del conjunto de datos porque aquí es suficiente para probar los datos con TSNE. El TSNE requiere demasiado tiempo para procesar, por lo tanto, usaré solo 3000 filas.

```python
!pip install tensorflow
```

```python
(x_train, y_train), (_ , _) = mnist.load_data()
x_train = x_train[:3000]
y_train = y_train[:3000]
print(x_train.shape) 
 
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[0], cmap = "gray")
```

MNIST es un dato tridimensional, lo transformaremos en uno bidimensional.

```python
x_mnist = reshape(x_train, [x_train.shape[0], x_train.shape[1]*x_train.shape[2]])
print(x_mnist.shape)
```

Aquí, tenemos 784 datos de características. Ahora, lo proyectaremos en dos dimensiones con TSNE y lo visualizaremos en un diagrama.

```python
tsne = TSNE(n_components=2, verbose=0, random_state=123)
z = tsne.fit_transform(x_mnist)
```

```python
#fit encontrar parametros (ubicación de los puntos)
#transform X -> X_transform, x_i nunca fue pasado al entranamiento 
#fit_transform fit + TransformedBbox

```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[5])
```

```python
reshape(x_train[5], [-1,784]) #reshape(x_train[5], [1,784]) #vector para poder tener un renglón
reshape(x_train[5], [-1,7]) #matriz de 112 renglones
```

```python
plt.figure(figsize=(20,90))
plt.imshow(reshape(x_train[5], [1, 784]))
```

```python
df = pd.DataFrame()
df["y"] = y_train
df["comp-1"] = z[:,0]
df["comp-2"] = z[:,1]
```

```python
sns.scatterplot(x="comp-1", y="comp-2", hue=df.y.tolist(),
                palette=sns.color_palette("hls", 10),
                data=df).set(title="MNIST data T-SNE projection")
```

El gráfico muestra una visualización bidimensional de los datos del MNIST. Los colores definen los dígitos objetivo y su ubicación de datos de características en el espacio 2D.
